```python
def sigmoid(x):
    return 1/(1+np.exp(-1*x))
sigmoid_matrix=np.vectorize(sigmoid)
```

```python
def Logistic_function(x,w,b):
    m=x.shape[0]
    #bnew=np.tile(b,(m,1))
    z=(np.matmul(x,w)+b)
    sigmoid_z=sigmoid_matrix(z)
    return sigmoid_z
```

```python
def cost_function(x,y,w,b):
    shape=x.shape
    m=shape[0]
    y_predicted=Logistic_function(x,w,b)
    loss=(((y*np.log(y_predicted)+((1-y)*np.log(1-y_predicted)))))
    cost=np.sum(loss,axis=0)
    return cost/-m
```

```python
def compute_gradient_mat(x,y,w,b):
    m=x.shape[0]
    dw=np.zeros(w.shape)
    db=np.zeros(b.shape)
    y_predicted=Logistic_function(x,w,b)
    loss=y_predicted-y
    dw=(1/m)*(np.matmul(x.T,loss))
    db=(1/m)*(np.sum(loss,axis=0))
    return dw,db               
```

```python
def gradient_decent(x,y,w,b,alpha,iterations):
    w_new=w
    b_new=b
    for i in range(iterations):
        dJ_w,dJ_b=compute_gradient_mat(x,y,w_new,b_new)
        w_new-=(alpha)*(dJ_w)
        b_new-=(alpha)*(dJ_b)
    return w_new,b_new
```
